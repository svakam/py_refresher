# Merging

```python
import pandas as pd
from pythonjsonlogger.core import merge_record_extra

left = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                     'A': ['A0', 'A1', 'A2', 'A3'],
                     'B': ['B0', 'B1', 'B2', 'B3']})
   
right = pd.DataFrame({'key': ['K0', 'K1', 'K2', 'K3'],
                          'C': ['C0', 'C1', 'C2', 'C3'],
                          'D': ['D0', 'D1', 'D2', 'D3']})   
```

Merge the dataframes `left` and `right` on their shared key.

```python
df_merged = left.merge(right, on='key')
df_merged
```

```python
left = pd.DataFrame({'key1': ['K0', 'K0', 'K1', 'K2'],
                     'key2': ['K0', 'K1', 'K0', 'K1'],
                        'A': ['A0', 'A1', 'A2', 'A3'],
                        'B': ['B0', 'B1', 'B2', 'B3']})
    
right = pd.DataFrame({'key1': ['K0', 'K1', 'K1', 'K2'],
                               'key2': ['K0', 'K0', 'K0', 'K0'],
                                  'C': ['C0', 'C1', 'C2', 'C3'],
                                  'D': ['D0', 'D1', 'D2', 'D3']})
```

```python
left
```

```python
right
```

Sometimes we will want to merge on more than one key. Try merging these two new dataframes by passing in a list of common keys to the `on` argument. Start with an inner join, and explain the output. Which values of `key1` and `key2` are retained? Which are dropped?

```python
merged = left.merge(right, on=['key1', 'key2'])
merged
```

Answer: The values of key1 and key2 that were retained are the combinations of both that are common to both 'left' and 'right'.

Now do the same thing, but with an outer join, left join, and right join. Discuss each. 

```python
merged_outer = left.merge(right, on=['key1', 'key2'], how='outer')
merged_outer
```

Answer: outer join shows results of both tables regardless of key1/key2 common combinations

```python
merged_left = left.merge(right, on=['key1', 'key2'], how='left')
merged_left
```

Answer: left join shows all results of 'left' table with common combinations with 'right', and any key1/key2 values from 'left' that don't appear in 'right'

```python
merged_right = left.merge(right, on=['key1', 'key2'], how='right')
merged_right
```

Answer: right join shows all results of 'right' table with common combinations with 'left', and any key1/key2 values from 'right' that don't appear in 'left'
